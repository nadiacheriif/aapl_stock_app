```python
# =========================================
# Stock Forecast Model Comparison Notebook
# =========================================

# STEP 0: Imports
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# ML models
from xgboost import XGBRegressor
from sklearn.preprocessing import MinMaxScaler

# Deep learning
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, LSTM, Dense, Input
from tensorflow.keras.optimizers import Adam

# Time series models
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet

```

```python
# =========================================
# STEP 1: Load Data
# =========================================
db_path = "/content/AAPL.db"  # Adjust path if needed
conn = sqlite3.connect(db_path)

# Load data (assuming table 'stock_prices')
df = pd.read_sql("SELECT * FROM stock_prices", conn, parse_dates=['Date'])
conn.close()

df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")
df = df.set_index("Date")

print(df.tail())

# Target variable
y = df["Close"].values

# Train/test split (80/20)
train_size = int(len(y) * 0.8)
train, test = y[:train_size], y[train_size:]

# =========================================
# STEP 2: Metrics function
# =========================================
def evaluate_model(name, actual, predicted):
    mae = mean_absolute_error(actual, predicted)
    rmse = sqrt(mean_squared_error(actual, predicted))
    mape = np.mean(np.abs((actual - predicted) / actual)) * 100
    return {"Model": name, "MAE": mae, "RMSE": rmse, "MAPE": mape}

results = []

```

```python
# =========================================
# STEP 3: ARIMA
# =========================================
try:
    model_arima = ARIMA(train, order=(5,1,0))
    model_fit = model_arima.fit()
    preds_arima = model_fit.forecast(steps=len(test))
    results.append(evaluate_model("ARIMA", test, preds_arima))
except Exception as e:
    print("ARIMA failed:", e)

```

```python
# =========================================
# STEP 4: Prophet
# =========================================
try:
    df_prophet = df.reset_index()[["Date", "Close"]]
    df_prophet.columns = ["ds", "y"]

    prophet = Prophet()
    prophet.fit(df_prophet.iloc[:train_size])

    future = prophet.make_future_dataframe(periods=len(test))
    forecast = prophet.predict(future)
    preds_prophet = forecast["yhat"].iloc[-len(test):].values
    results.append(evaluate_model("Prophet", test, preds_prophet))
except Exception as e:
    print("Prophet failed:", e)
```

```python
# =========================================
# STEP 5: XGBoost
# =========================================
try:
    X = np.arange(len(y)).reshape(-1, 1)
    X_train, X_test = X[:train_size], X[train_size:]

    model_xgb = XGBRegressor(n_estimators=200, random_state=42)
    model_xgb.fit(X_train, train)
    preds_xgb = model_xgb.predict(X_test)
    results.append(evaluate_model("XGBoost", test, preds_xgb))
except Exception as e:
    print("XGBoost failed:", e)
```

```python

# =========================================
# STEP 6: LSTM
# =========================================
try:
    lookback = 60
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled = scaler.fit_transform(y.reshape(-1,1))

    def create_sequences(data, lookback):
        X, Y = [], []
        for i in range(len(data) - lookback):
            X.append(data[i:i+lookback])
            Y.append(data[i+lookback])
        return np.array(X), np.array(Y)

    X_all, Y_all = create_sequences(scaled, lookback)
    X_train, X_test = X_all[:train_size-lookback], X_all[train_size-lookback:]
    Y_train, Y_test = Y_all[:train_size-lookback], Y_all[train_size-lookback:]

    model_lstm = Sequential([
        Input(shape=(lookback, 1)),
        LSTM(50, return_sequences=True),
        LSTM(50),
        Dense(1)
    ])
    model_lstm.compile(optimizer=Adam(0.001), loss="mse")
    model_lstm.fit(X_train, Y_train, epochs=10, batch_size=32, verbose=0)

    preds_scaled = model_lstm.predict(X_test)
    preds_lstm = scaler.inverse_transform(preds_scaled).flatten()
    Y_test_inv = scaler.inverse_transform(Y_test.reshape(-1,1)).flatten()

    results.append(evaluate_model("LSTM", Y_test_inv, preds_lstm))
except Exception as e:
    print("LSTM failed:", e)
```

```python
# =========================================
# STEP 7: GRU
# =========================================
try:
    model_gru = Sequential([
        Input(shape=(lookback, 1)),
        GRU(50, return_sequences=True),
        GRU(50),
        Dense(1)
    ])
    model_gru.compile(optimizer=Adam(0.001), loss="mse")
    model_gru.fit(X_train, Y_train, epochs=10, batch_size=32, verbose=0)

    preds_scaled = model_gru.predict(X_test)
    preds_gru = scaler.inverse_transform(preds_scaled).flatten()

    results.append(evaluate_model("GRU", Y_test_inv, preds_gru))
except Exception as e:
    print("GRU failed:", e)
```

```python
# =========================================
# STEP 8: Results Table
# =========================================
results_df = pd.DataFrame(results)
print(results_df)

# =========================================
# STEP 9: Visualization
# =========================================
plt.figure(figsize=(12,6))
plt.plot(df.index[train_size:], test, label="Actual", color="black")

if "preds_arima" in locals(): plt.plot(df.index[train_size:], preds_arima, label="ARIMA")
if "preds_prophet" in locals(): plt.plot(df.index[train_size:], preds_prophet, label="Prophet")
if "preds_xgb" in locals(): plt.plot(df.index[train_size:], preds_xgb, label="XGBoost")
if "preds_lstm" in locals(): plt.plot(df.index[train_size:], preds_lstm, label="LSTM")
if "preds_gru" in locals(): plt.plot(df.index[train_size:], preds_gru, label="GRU")

plt.title("Model Comparison: Actual vs Predicted")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import random
from collections import deque
import matplotlib.pyplot as plt

# ======================
# 1. Trading Environment
# ======================
class TradingEnv:
    def __init__(self, df, window_size=10, initial_balance=10000):
        self.df = df.reset_index(drop=True)
        self.window_size = window_size
        self.initial_balance = initial_balance
        self.reset()

    def reset(self):
        self.current_step = self.window_size
        self.balance = self.initial_balance
        self.shares_held = 0
        self.total_profit = 0
        self.done = False
        return self._get_state()

    def _get_state(self):
        # Use last `window_size` returns as state
        window_data = self.df['Close'][self.current_step - self.window_size:self.current_step].values
        norm_window = window_data / window_data[0] - 1.0  # normalize
        return np.concatenate(([self.balance, self.shares_held], norm_window))

    def step(self, action):
        # Actions: 0 = Hold, 1 = Buy, 2 = Sell
        price = self.df['Close'][self.current_step]
        reward = 0

        if action == 1:  # Buy
            if self.balance > price:
                self.shares_held += 1
                self.balance -= price
        elif action == 2:  # Sell
            if self.shares_held > 0:
                self.shares_held -= 1
                self.balance += price
                reward = price - self.df['Close'][self.current_step - 1]  # profit

        self.current_step += 1
        if self.current_step >= len(self.df):
            self.done = True

        self.total_profit = self.balance + self.shares_held * price - self.initial_balance
        return self._get_state(), reward, self.done

# ==========================
# 2. Deep Q-Learning Agent
# ==========================
class DQNAgent:
    def __init__(self, state_size, action_size, lr=0.001, gamma=0.95, epsilon=1.0, epsilon_min=0.01, epsilon_decay=0.995):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        self.gamma = gamma
        self.epsilon = epsilon
        self.epsilon_min = epsilon_min
        self.epsilon_decay = epsilon_decay
        self.model = self._build_model(lr)

    def _build_model(self, lr):
        model = nn.Sequential(
            nn.Linear(self.state_size, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, self.action_size)
        )
        self.optimizer = optim.Adam(model.parameters(), lr=lr)
        self.loss_fn = nn.MSELoss()
        return model

    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        if np.random.rand() <= self.epsilon:
            return random.randrange(self.action_size)
        state_tensor = torch.FloatTensor(state).unsqueeze(0)
        q_values = self.model(state_tensor)
        return torch.argmax(q_values).item()

    def replay(self, batch_size=32):
        if len(self.memory) < batch_size:
            return
        minibatch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in minibatch:
            state_tensor = torch.FloatTensor(state).unsqueeze(0)
            next_state_tensor = torch.FloatTensor(next_state).unsqueeze(0)
            target = reward
            if not done:
                target = reward + self.gamma * torch.max(self.model(next_state_tensor)).item()
            target_f = self.model(state_tensor)
            target_f = target_f.clone()
            target_f[0][action] = target
            output = self.model(state_tensor)
            loss = self.loss_fn(output, target_f)
            self.optimizer.zero_grad()
            loss.backward()
            self.optimizer.step()
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

# ==========================
# 3. Training Loop
# ==========================
env = TradingEnv(df, window_size=10)  # df should be your stock dataframe
state_size = len(env.reset())
action_size = 3
agent = DQNAgent(state_size, action_size)

episodes = 50
profits = []

for e in range(episodes):
    state = env.reset()
    total_reward = 0
    while not env.done:
        action = agent.act(state)
        next_state, reward, done = env.step(action)
        agent.remember(state, action, reward, next_state, done)
        state = next_state
        total_reward += reward
        agent.replay()
    profits.append(env.total_profit)
    print(f"Episode {e+1}/{episodes}, Total Profit: {env.total_profit:.2f}, Epsilon: {agent.epsilon:.2f}")

# ==========================
# 4. Results Visualization
# ==========================
plt.plot(profits)
plt.xlabel("Episode")
plt.ylabel("Total Profit")
plt.title("RL Trading Agent Performance")
plt.show()
```
